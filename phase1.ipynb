```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset(portfolio dataset)
df = pd.read_csv('/content/portfolio_data.csv')

# Convert 'Date' column to datetime format and set it as the index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Check the first few rows and data types
print(df.head())
print(df.dtypes)

# Check for missing values and handle them if necessary
print(df.isnull().sum())

# Plot adjusted closing prices for each stock
plt.figure(figsize=(10, 6))
plt.plot(df['AMZN'], label='Amazon (AMZN)')
plt.plot(df['DPZ'], label='Domino\'s Pizza (DPZ)')
plt.plot(df['BTC'], label='Bitcoin (BTC)')
plt.plot(df['NFLX'], label='Netflix (NFLX)')
plt.title('Adjusted Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

```
